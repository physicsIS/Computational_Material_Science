```python
import matplotlib.pyplot as plt
from matplotlib.patches import  FancyArrowPatch
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
Img = plt.imread('frame0_x.png')
ax.imshow(Img)



ax.annotate('Carbon Nanoscroll', xy=(475, 50), xytext=(525, 20),
			arrowprops=dict(facecolor='white', shrink=0.05))
ax.annotate('Ionic Liquid Electrolyte \n' + r"AlCl$_3$/[Emim]Cl", xy=(500, 250), xytext=(550, 225),
			arrowprops=dict(facecolor='white', shrink=0.05))

# Bracket
center = ((15,125), (40,125))
angleAB = (0,0)
style = "-["
bracketstyle = f"{style}, angleB ={angleAB[0]}"
bracket = FancyArrowPatch(*center, arrowstyle=bracketstyle,mutation_scale=20)
ax.add_patch(bracket)

ax.annotate('Fixed Carbons', xy=(27.5, 125), xytext=(-25, 130),
			ha='center', va='bottom')


fig.savefig('interface_fixed_carbons.pdf', dpi=300, bbox_inches='tight')
```

```python
from ase.io.espresso import read_espresso_in
from ase.io import read, write

# Lee archivo de entrada
atoms_in = read_espresso_in('inputs/pw_inicial.in')

# Guarda restricciones (flags if_pos)
constraints = getattr(atoms_in, 'constraints', None)
```

```python
# Lee la última estructura del archivo de salida
atoms_out = read('outputs_3/pw.out.545', format='espresso-out')
```

```python
if constraints is not None:
    atoms_out.set_constraint(constraints)
```

```python
write("inputs/pw_new3.in", atoms_out, format="espresso-in", 
	pseudopotentials={
	"C": "C.pbe-n-kjpaw_psl.1.0.0.UPF",  # Especifica el pseudopotencial para carbono.
	"H": "H.pbe-kjpaw_psl.1.0.0.UPF",     # Especifica el pseudopotencial para hidrógeno.
	"N": "N.pbe-n-kjpaw_psl.1.0.0.UPF",
	"Al": "Al.pbe-n-kjpaw_psl.1.0.0.UPF",
	"Cl": "Cl.pbe-n-kjpaw_psl.1.0.0.UPF"
})
```
